```python
import numpy as np 
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_excel("App Analytics Report-06.05.2023.xlsx",sheet_name="User Acquisition")
```

```python
df1
```

```python
df2=pd.read_excel("App Analytics Report-06.05.2023.xlsx",sheet_name="Traffic Aquisition")
```

```python
df2
```

```python
# Plotting New Users and Engaged Sessions for both datasets
plt.figure(figsize=(15, 4))

# New Users
plt.subplot(1, 2, 1)
sns.barplot(x="First user default channel group", y="Engaged sessions", data=df1)
plt.title("Engaged Sessions - df1")

# Engaged Sessions
plt.subplot(1, 2, 2)
sns.barplot(x="Session default channel group", y="Engaged sessions", data=df2)
plt.title("Engaged Sessions - df2")

plt.tight_layout()
plt.show()
```

```python
# Plotting New Users and Engaged Sessions for both datasets
plt.figure(figsize=(15, 4))

# New Users
plt.subplot(1, 2, 1)
sns.barplot(x="First user default channel group", y="Average engagement time", data=df1)
plt.title("Engaged Sessions - df1")

# Engaged Sessions
plt.subplot(1, 2, 2)
sns.barplot(x="Session default channel group", y="Average engagement time per session", data=df2)
plt.title("Engaged Sessions - df2")

plt.tight_layout()
plt.show()
```
